Because Chol is around 300 and Oldpeak is around 1, the "lbfgs" solver is having a hard time navigating the math. It's like trying to walk down a hill where one step is 1 meter long and the next step is 300 meters long—it's clumsy and slow.

If you scale your data using StandardScaler, this error will likely disappear even with the default 100 iterations.
